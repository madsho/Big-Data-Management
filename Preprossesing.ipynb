```python
import pandas as pd
import numpy as np
from collections import Counter
```

# Import and read data


```python
df = pd.read_csv('SBAnational.csv', sep=",")
df.head(5)
```

# Drop irrelevant attributes


```python
df = df.drop(columns=['LoanNr_ChkDgt','Name', 'ChgOffDate','DisbursementDate','BalanceGross','ChgOffPrinGr'])

df.info()
```

# Filter for dates (reducing dataset)

```python
df['ApprovalFY'] = pd.to_numeric(df['ApprovalFY'], errors='coerce').astype(pd.Int64Dtype())

df.info()
```

```python
df = df.drop(df[(df['ApprovalFY'] < 2000) | (df['ApprovalFY'] > 2010)].index)

df.info()
```

```python
df.dropna(inplace=True)
df.head()
```

Process remaining attributes

# City

```python
countCity = Counter(df['City'])
print(countCity)
```

# NAICS

```python
# astype(str) is used to convert the values in the columns to strings, and then str[:2] is used to retain only the first two characters of each string. Finally, the modified values are assigned back to the respective columns in the DataFrame.

df['NAICS'] = df['NAICS'].astype(str).str[:2]
df['NAICS'] = df['NAICS'].astype(int)
df.info()
```

# Newexist

```python
df['NewExist'] = df['NewExist'].astype(int)
countNewExist = Counter(df['NewExist'])
print(countNewExist)
```

```python
df.info()
```

```python
df = df.loc[df['NewExist'] != 0]
countNewExist = Counter(df['NewExist'])
print(countNewExist)
```

# Franchisecode

```python
countFranchiseCode = Counter(df['FranchiseCode'])
print(countFranchiseCode)
```

```python
df['FranchiseCode'] = np.where(df['FranchiseCode'] == 1, 0, df['FranchiseCode'])
df['FranchiseCode'] = np.where(df['FranchiseCode'] > 0, 1, df['FranchiseCode'])
countFranchiseCode = Counter(df['FranchiseCode'])
print(countFranchiseCode)
```

# RevLineCr

```python
countRevLineCr = Counter(df['RevLineCr'])
print(countRevLineCr)
```

# Lowdoc

```python
countLowDoc = Counter(df['LowDoc'])
print(countLowDoc)
```

```python
df = df[df['LowDoc'].isin(['N', 'Y'])]
countLowDoc = Counter(df['LowDoc'])
print(countLowDoc)
```

```python
df['LowDoc'] = np.where(df['LowDoc'] == 'Y', 1, df['LowDoc'])
df['LowDoc'] = np.where(df['LowDoc'] =='N', 0, df['LowDoc'])
countLowDoc = Counter(df['LowDoc'])
print(countLowDoc)
```

```python
df.info()
```

# Disburstmengross

```python
print(df['DisbursementGross'].values)
```

```python
df['DisbursementGross'] = df['DisbursementGross'].replace('[\$,]', '', regex=True).astype(float)
print(df['DisbursementGross'].values)
```

```python
countDisbursementGross = Counter(df['DisbursementGross'])
print(countDisbursementGross)
```

```python
contains_zero = (df['DisbursementGross'] == 0).any()
print(contains_zero)
```

# Urban_Rural

```python
countUrbanRural = Counter(df['UrbanRural'])
print(countUrbanRural)
```

```python
df = df.loc[df['UrbanRural'] != 0]
countUrbanRural = Counter(df['UrbanRural'])
print(countUrbanRural)
```

```python
df.info()
```

# MIS_STATUS

```python
countMIS_Status = Counter(df['MIS_Status'])
print(countMIS_Status)
```

```python
df['MIS_Status'] = np.where(df['MIS_Status'] == 'P I F', 1, df['MIS_Status'])
df['MIS_Status'] = np.where(df['MIS_Status'] =='CHGOFF', 0, df['MIS_Status'])
countMIS_Status = Counter(df['MIS_Status'])
print(countMIS_Status)
```

# Ratio

### Clean Data

```python
print(df['GrAppv'].values)
print(df['SBA_Appv'].values)
```

```python
df['GrAppv'] = df['GrAppv'].replace('[\$,]', '', regex=True).astype(float)
countGrAppv = Counter(df['GrAppv'])

df['SBA_Appv'] = df['SBA_Appv'].replace('[\$,]', '', regex=True).astype(float)
countSBA_Appv = Counter(df['SBA_Appv'])

print(countGrAppv)
print(countSBA_Appv)
```

```python
contains_zero = (df['GrAppv'] == 0).any()
contains_zeroSBA_Appv = (df['SBA_Appv'] == 0).any()
print(contains_zero)
print(contains_zeroSBA_Appv)
```

### Create Ratio Column

```python
df['Ratio'] = df['SBA_Appv'] / df['GrAppv']
print(df['Ratio'].values)
```

```python
countRatio = Counter(df['Ratio'])
print(countRatio)
```

### Remove the SBA_Appv and GrAppv

```python
df = df.drop(columns=['SBA_Appv', 'GrAppv'])

df.info()
```

### Processed data is loaded to CSV

```python
df.to_csv('PreProcessedData.csv')
```
